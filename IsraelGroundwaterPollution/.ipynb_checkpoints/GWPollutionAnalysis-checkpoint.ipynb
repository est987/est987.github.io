```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
df = pd.read_csv('C:/Users/eliav/Documents/Projects/IsraelGroundwaterPollution/data/wa_aquifer_pol_conc_02_18.csv',
           encoding='windows-1255')
#df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={df.columns[0] : "id",
                       df.columns[1] : "name",
                       df.columns[2] : "cell", #aquifer_cell
                       df.columns[3] : "op", #operator
                       df.columns[4] : "pol", #pollutant
                       df.columns[5] : "st", #standard_type
                       df.columns[6] : "sv", #standard_value
                       df.columns[7] : "date",#sample_date
                       df.columns[8] : "c", #concentration
                       df.columns[9] : "pct"}) #pct_of_standard
df=df.iloc[1:,:]
```

```python
#df.head()
```

```python
#drop hebrew column names
#set numeric types
df.iloc[:,[0,2,6,8,9]] = df.iloc[:,[0,2,6,8,9]].apply(pd.to_numeric)
df.loc[:,'date'] = pd.to_datetime(df.date, dayfirst=True)
df[['cell','st','op']] = df[['cell','st','op']].astype('category')

df.date.head()
```

```python
df['year'] = df['date'].dt.year
df.sort_values('year',inplace=True)
df['year'] = df['year'].astype('category')
df['year'].unique()
```

```python
df.dtypes
```

```python
#Cl in NO3 only
dfp = df.loc[df['pol'].isin(['Cl','NO3'])]

dfp.drop(columns='sv',inplace=True)
dfp[['pol']] = dfp[['pol']].astype('category')
#dfp.head()
#dfp.dtypes
```

```python
df2 = dfp
df3 = df2.groupby(['cell','pol','year'],observed=True).mean()
#df3.index.get_level_values(1).unique()
#df3 = df3.reset_index()
#df2.plot(y="pct_of_standard")
```

```python
df3 = df2.pivot_table(index=['id','cell'] ,columns='pol', aggfunc=np.mean)
df3.head()
```
